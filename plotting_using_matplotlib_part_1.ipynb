```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#F5E427')
plt.plot(batsman['index'],batsman['RG Sharma'], color='#F52749')

plt.title('Rohit Sharma vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#F5E427',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'], color='#F52749',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#F5E427',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'], color='#F52749',linestyle='dashdot',linewidth=2,marker='+')

plt.title('Rohit Sharma vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legends -> Location
plt.plot(batsman['index'],batsman['V Kohli'],color='#F5E427',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color='#F52749',linestyle='dashdot',linewidth=2,marker='+',label='Rohit')

plt.title('Rohit Sharma vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='best')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2016,2017,2018,2019,2020,2021,2021]

plt.plot(year,price)
plt.ylim(0,60000)
plt.xlim(2016,2019)
```

# Scatter Plots

image.png

* Bivariate Analysis
* numerical vs numerical
* Use case- Finding correlation

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('C:/Users/Ranjan Singh/Downloads/all_pandas_data_sets/datasets-7/batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike_Rate')
```

```python
# marker
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='+')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike_Rate')
```

```python
# size
tips = sns.load_dataset('tips')

# slower technique
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
# scatterplot using plt.plot
# faster technique
plt.plot(tips['total_bill'],tips['tip'],'o')
```

# plt.plot vs plt.scatter
plt.plot -> it's  also faster on huge datasets but we cannot change size, color on it.<br>
plt.scatter -> it's slower on huge datasets but we can change color, size on it.
