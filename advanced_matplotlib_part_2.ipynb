```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Colored Scatterplots

```python
iris = pd.read_csv('C:/Users/Ranjan Singh/Downloads/all_pandas_data_sets/datasets-8/iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
# cmap and alpha
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.9)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

# plot size

```python
plt.figure(figsize=(15,7))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.9)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
batters = pd.read_csv('C:/Users/Ranjan Singh/Downloads/all_pandas_data_sets/datasets-8/batter.csv')
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# Horizontal and Vertical lines

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# Subplots

```python
# A diff way to plot graphs
batters.head()
```

```python
plt.figure(figsize=(15,6))

plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('somethings')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='D')
ax.set_title('somethings')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
# batter dataset
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red',marker='D')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# This is good way to use subplots

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red',marker='D')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
# In this way also we can use subplots

fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red',marker='D')

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(batters['avg'],batters['runs'],color='green',marker='+')

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])

ax4 = fig.add_subplot(2,2,4)
ax4.hist(batters['runs'])
```
