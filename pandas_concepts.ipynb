# What is Pandas
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.

https://pandas.pydata.org/about/index.html

### Pandas Series
A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### Importing Pandas

```python
import numpy as np
import pandas as pd
```

### Series from lists

```python
# string
country = ["India", "USA", "UK", "Germany", "France"]
pd.Series(country)
```

```python
# Integers
runs = [10, 15, 20, 50, 100]

runs_series = pd.Series(runs)
runs_series
```

```python
# cunstom index
marks = [67, 57, 82, 90]
subjects = ['maths', 'english', 'science', 'social']

pd.Series(marks, index=subjects)
```

```python
# setting a name to the series
marks = pd.Series(marks,index=subjects,name='Ranjan marks')
marks
```

### Series from dict

```python
marks = {
    'maths':100,
    'english':90,
    'science':87,
    'social':78
}

marks_series = pd.Series(marks,name='Ranjan marks')
marks_series
```
